```python
import pandas as pd
df = pd.read_csv('./deliveries.csv')
df.head()
df.dtypes
```

```python
df.head()
```

```python
# Cast the relevant columns to their appropriate types
df['batsman_runs'] = df['batsman_runs'].astype(int)
df['extra_runs'] = df['extra_runs'].astype(int)
df['total_runs'] = df['total_runs'].astype(int)
df['is_wicket'] = df['is_wicket'].astype(int)

# Calculate total runs scored by each batsman
batsman_runs = df.groupby('batter')['batsman_runs'].sum().reset_index().rename(columns={'batsman_runs': 'total_runs'})
```
